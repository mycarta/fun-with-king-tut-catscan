#### Import libraries

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from skimage import io, feature, color
from skimage.filters import sobel
%matplotlib inline
```

#### Import image - cat scan of King Tut skull (Supreme Council of Antiquities)
From http://guardians.net/hawass/press_release_tutankhamun_ct_scan_results.htm

Read more at http://www.two-views.com/celebrity/king-tut-xray.html#sthash.YuCwv0p2.dpbs


```python
# t=io.imread('http://www.two-views.com/images/Tut%20bone%20frag.jpg') 
# can't seem to open it directly from url http://www.two-views.com/images/Tut%20bone%20frag.jpg

t=io.imread('Tut.jpg')  
tut = color.rgb2gray(t) 


print tut.shape

fig = plt.figure(figsize=(8,8))

ax = fig.add_subplot(1, 1, 1)
ax.set_xticks([])
ax.set_yticks([])

plt.imshow(tut, cmap='gray')
plt.show()
```

#### Decimate by a factor of 4 - simulates relatively high density seismic acquisition and run edge detection to simulate discontinuity attribute

```python
tut_r1 = tut[::4,::4]
#edges_r1 = feature.canny(tut_r1)
edges_r1 = sobel(tut_r1)

print tut_r1.shape

fig = plt.figure(figsize=(20,20))

ax3 = fig.add_subplot(1, 2, 1)
plt.imshow(tut_r1, cmap='gray')
ax3.set_xticks([])
ax3.set_yticks([])

ax4 = fig.add_subplot(1, 2, 2)
plt.imshow(edges_r1, cmap='gray')
ax4.set_xticks([])
ax4.set_yticks([])

plt.show()

```

#### Additional decimation by a factor of 2 - simulates medium density seismic acquisition, followed by interpolation to upsample to original size, and run edge detection to simulate discontinuity attribute

```python
# - simulates medium density seismic acquisition

tut_i0 = tut_r1[::2,::2]
tut_i1 = sp.ndimage.interpolation.zoom(tut_i0, 2, order=2)
edges_i1 = sobel(tut_i1)

print tut_r1.shape

fig = plt.figure(figsize=(20,20))

ax3 = fig.add_subplot(1, 2, 1)
plt.imshow(tut_i1, cmap='gray')
ax3.set_xticks([])
ax3.set_yticks([])

ax4 = fig.add_subplot(1, 2, 2)
plt.imshow(edges_i1, cmap='gray')
ax4.set_xticks([])
ax4.set_yticks([])

plt.show()
```

#### Additional  decimation by a factor of 2 in one direction and 4 in the other, followed by interpolation to upsample to original size, and   run edge detection to simulate discontinuity attribute.  Very deteriorated edge result. Can this be considered a simulation of megabin seismic acquisition and 5D interpolation? Probably not, because images do not carry any azimuthal information. This is more like a post-stack decimation followed by interpolation.

```python
# further reduction by a factor of 2 in one direction and 4 in the other - simulates megabin seismic acquisition
tut_i00 = tut_r1[::2,::4]
tut_i11 = sp.ndimage.interpolation.zoom(tut_i00, (2,4), order=2)
edges_i11 = sobel(tut_i11)

print tut_i11.shape

fig = plt.figure(figsize=(20,20))

ax3 = fig.add_subplot(1, 2, 1)
plt.imshow(tut_i11, cmap='gray')
ax3.set_xticks([])
ax3.set_yticks([])

ax4 = fig.add_subplot(1, 2, 2)
plt.imshow(edges_i11, cmap='gray')
ax4.set_xticks([])
ax4.set_yticks([])

plt.show()
```
